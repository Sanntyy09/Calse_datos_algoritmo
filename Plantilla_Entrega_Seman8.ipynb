## 1) Descripción del problema
En este laboratorio se debe desarrollar un sistema para gestionar información utilizando estructuras de datos tipo árbol.

La información se organiza mediante códigos únicos, los cuales permiten identificar cada elemento dentro del sistema. Para esto, se emplean árboles binarios de búsqueda (BST), que permiten insertar y buscar datos de manera eficiente siguiendo un orden lógico.

También se estudia el problema del desbalanceo en los árboles y se introduce el uso de árboles AVL, los cuales mantienen el equilibrio automáticamente para garantizar un mejor rendimiento en las operaciones.

El sistema debe permitir insertar datos, buscarlos, recorrer el árbol y analizar su estructura.

```python
# Escribe tu código aquí.
# Debe incluir: validación, función memoizada, y ejecución.
# verifica si es numero entero positivo
 
def es_numero(texto):
    if texto == "":
        return False
    i = 0
    while i < len(texto):
        if texto[i] < "0" or texto[i] > "9":
            return False
        i = i + 1
    return True
# lee opcion valida del menu
def leer_opcion():
    while True:
        entrada = input("seleccione opcion: ").strip()
        if entrada == "":
            print("ingrese una opcion valida (1-5)")
            continue
        # si hay espacios, tomamos el primer token
        if " " in entrada:
            entrada = entrada.split()[0]
        if es_numero(entrada):
            opcion = int(entrada)
            if 1 <= opcion <= 5:
                return opcion
            else:
                print("opcion invalida, ingrese 1-5")
        else:
 
            print("ingrese numeros positivos solamente")
# lee lista de numeros
def leer_lista(mensaje):
    while True:
        texto = input(mensaje)
        partes = texto.split()
        if len(partes) == 0:
            print("ingrese numeros positivos solamente")
        else:
            lista = []
            valido = True
            i = 0
            while i < len(partes):
                if es_numero(partes[i]):
                    lista.append(int(partes[i]))
                else:
                    valido = False
                    break
                i = i + 1
            if valido:
                return lista
            else:
                print("ingrese numeros positivos solamente")
# nodo bst: [valor, hijo_izq, hijo_der]
def crear_nodo_bst(numero):
    return [numero, [], []]
# inserta un valor en el bst
def insertar_bst(arbol, numero):
    if arbol == []:
        return crear_nodo_bst(numero)
    if numero < arbol[0]:
        arbol[1] = insertar_bst(arbol[1], numero)
    elif numero > arbol[0]:
        arbol[2] = insertar_bst(arbol[2], numero)
    return arbol
# busca un numero en el bst
def buscar_bst(arbol, numero):
    if arbol == []:
        return False 
    if numero == arbol[0]:
        return True
    if numero < arbol[0]:
        return buscar_bst(arbol[1], numero)
    else:
        return buscar_bst(arbol[2], numero)
# recorrido preorden
def recorrido_preorden(arbol, lista):
    if arbol != []:
        lista.append(arbol[0])
        recorrido_preorden(arbol[1], lista)
        recorrido_preorden(arbol[2], lista)
# recorrido inorden
def recorrido_inorden(arbol, lista):
    if arbol != []:
        recorrido_inorden(arbol[1], lista)
        lista.append(arbol[0])
        recorrido_inorden(arbol[2], lista)
# recorrido postorden
def recorrido_postorden(arbol, lista):
    if arbol != []:
        recorrido_postorden(arbol[1], lista)
        recorrido_postorden(arbol[2], lista)
        lista.append(arbol[0])
# muestra el arbol
def mostrar_arbol(arbol, nivel, lado):
    if arbol != []:
        print("   " * nivel + lado + ": " + str(arbol[0]))
        mostrar_arbol(arbol[1], nivel + 1, "izq")
        mostrar_arbol(arbol[2], nivel + 1, "der")
# construye bst desde lista
def construir_bst(lista):
    arbol = []
    i = 0
    while i < len(lista):
        arbol = insertar_bst(arbol, lista[i])
        i = i + 1
    return arbol
# actividad 1
def actividad_1():
    lista = leer_lista("ingrese los codigos para construir el bst: ")
    arbol = construir_bst(lista)
    print("arbol bst construido")
    mostrar_arbol(arbol, 0, "raiz")
    return arbol
# actividad 2
def actividad_2(arbol):
    if arbol == []:
        print("primero debe construir el bst en la actividad 1")
    else:
        lista = leer_lista("ingrese los valores a buscar: ")
        i = 0
        while i < len(lista):
            if buscar_bst(arbol, lista[i]):
                print("buscar", lista[i], ": encontrado")
            else:
                print("buscar", lista[i], ": no encontrado")
            i = i + 1
# actividad 3
def actividad_3(arbol):
    if arbol == []:
        print("primero debe construir el bst en la actividad 1")
    else:
        pre = []
        ino = []
        post = []
        recorrido_preorden(arbol, pre)
        recorrido_inorden(arbol, ino)
        recorrido_postorden(arbol, post)
        print("preorden:", pre)
        print("inorden:", ino)
        print("postorden:", post)
# actividad 4
def actividad_4(arbol):
    if arbol == []:
        print("primero debe construir el bst en la actividad 1")
    else:
        lista = leer_lista("ingrese los nuevos valores a insertar: ")
        i = 0
        while i < len(lista):
            arbol = insertar_bst(arbol, lista[i])
            print("despues de insertar", lista[i])
            mostrar_arbol(arbol, 0, "raiz")
            i = i + 1
    return arbol
# menu
def mostrar_menu():
    print()
    print("menu")
    print("1 actividad 1 construccion del arbol bst")
    print("2 actividad 2 busqueda")
    print("3 actividad 3 recorridos")
    print("4 actividad 4 insercion de nuevos computadores")
    print("5 salir")
# programa principal
def main():
    arbol_bst = []
    opcion = 0
    while opcion != 5:
        mostrar_menu()
        opcion = leer_opcion()
        if opcion == 1:
            arbol_bst = actividad_1()
        elif opcion == 2:
            actividad_2(arbol_bst)
        elif opcion == 3:
            actividad_3(arbol_bst)
        elif opcion == 4:
            arbol_bst = actividad_4(arbol_bst)
        elif opcion == 5:
            print("fin del programa")
if __name__ == '__main__':
    main()
```
